# Introduction

The LoLN tells us two things :

1. The average of many independent samples is with high probability close to the mean of the underlying distribution.
2. The density histogram of many independent samples is with high probability close to the graph of the density of the underlying distribution.

The CLT states that the sum or average of many independent copies of a random variable is approximately a normal random variable. The CLT goes on to give the precise values for the mean
and standard deviation of the normal random variable.


## Law of Large Numbers

Suppose $X_1,X_2, \cdots, X_n$ are independent random variables with the same underlying distribution. In this case, we say that the $X_i$ are independent and identically distributed or i.i.d. In particular, the $X_i$, all have the same mean $\mu$ and standard deviation $\sigma$.

Let $\bar{X_n}$ be the average of $N$ independent samples of $X_i$:
$$ \bar{X*N} = \frac{1}{N} \sum*{i=1}^N X_i $$

The Law of Large Numbers states that as $n$ grows, the probability that $\bar{X_n}$ is close to $\mu$ is close to 1.

The Central Limit Theorem states that as $n$ grows, the distribution of $\bar{X_n}$ converges to the normal distribution $ \mathcal{N}(\mu, {\sigma^2}/{n})$.

# [20 points] Task 03 - Central Limit Theorem


In this task, you will verify the CLT i.e. distrubution of $\bar{X_N}$ converges to a normal distribution $\mathcal{N} (\mu, \sigma^2/n) $ for both the discrete and continous case. In the continous case, you will be working with exponential and uniform random variables. In the discrete case, you will be working with Bernoulli random variables.


## [5 points] Part A: Uniform Random Variable

In this part, you are required to complete the function uniform_clt(sample_size,population_size) which computes the normal approximation for the uniform random variables. Furthermore, you are required to comment of the how the shape of the graph changes as you change the value of $n$. Each ${X_i} \sim U(0,1)$


```python
# TODO: Complete function uniform_clt
def uniform_clt(sample_size, population_size):
    """
    This function computes the normal approximation for the uniform random variables.
    """
    # Mean of the uniform random variable
    mean = 1 / 2
    # Standard deviation of the uniform random variable
    std = 12 ** (-0.5)
    # Create Dataframe to store sampling results
    df = pd.DataFrame()
    # Simulation
    for i in range(population_size):
        # Generate uniform random variables
        uniform_sample = np.random.uniform(0, 1, sample_size)
        col = f"sample {i}"
        df[col] = uniform_sample
    # Compute sample means
    df_sample = pd.DataFrame(df.mean(), columns=["Sample means"])
    # X_n ~ N(mu, sigma^2/2)
    clt_mu = mean
    clt_sigma = std / np.sqrt(sample_size)
    s_mu = df_sample.mean().values[0]
    s_sigma = df_sample.std().values[0]
    # Plot distribution of sample means
    sns.distplot(df_sample, hist=False, label="Actual Data")
    # Plot CLT distribution
    X = np.linspace(clt_mu - 3 * clt_sigma, clt_mu + 3 * clt_sigma, 100)
    Y = stats.norm.pdf(X, clt_mu, clt_sigma)
    plt.plot(X, Y, label="Normal approximation")
    plt.title(f"Sample size = {sample_size}")
    plt.legend()
    plt.ylabel("Density")
    plt.xlabel("x")
    plt.show()
    # Print CLT statistics vs actual statistics
    print(f"CLT mu = {clt_mu} and CLT sigma = {clt_sigma}")
    print(f"Sample mu = {s_mu} and Sample sigma = {s_sigma}")


for i in [1, 10, 50, 200, 500, 1000, 10000, 100000]:
    uniform_clt(i, 100)
```
